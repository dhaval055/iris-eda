```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```
df = pd.read_csv('/content/Iris.csv')
df.head()
```

```
del df['Id']
```

```
df.info()
```

```
df.describe()
```

```
df['Species'].value_counts()
```

**Takeaway :-**
- Data is perfectly balanced.

```
sns.FacetGrid(df,hue="Species",size=4).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
plt.show()
```

**Takeaway :-**
- setoesa is linearly seprable.
- versicolor and virginica are hard to seperate with sepal   length and width. 

```
sns.FacetGrid(df,hue="Species",size=4).map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend()
plt.show()
```

**Takeaway :-**
- Petal length and width makes much more easy to seperate flowers.

```
sns.pairplot(df,hue="Species",height=3);
plt.show()
```

**Takeaway :-**
- Petal width and length seems to be most useful.
- Setosa is linearly seperable, versicolor and virginica are almost linearly sperable.


```
sns.FacetGrid(df,hue="Species",size=5).map(sns.distplot,"PetalWidthCm").add_legend();    
plt.show()
```

Takeaway :-
- There is a overlap between versicolor and virginica.

```
sns.FacetGrid(df,hue="Species",size=5).map(sns.distplot,"PetalLengthCm").add_legend();    
plt.show()
```

Takeaway :-
- There is a overlap between versicolor and virginica.

```
sns.boxplot(data=df,x="Species",y="PetalLengthCm")
plt.show()
```

**Conclusion:-** 
- PetalLength and PetalWidth are most useful features.
- Setosa is easily seperable from other two.
- Versicolor and virginica are almost linearly seperable.
- Overlap in petal width is lesser than petal length for all three categories.
